```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

data= {
    "size": [500,1000,1500,2000,2500],
    "price": [100,200,300,400,500]
}

df= pd.DataFrame(data)

x= df[["size"]]
y= df["price"]

model= LinearRegression()
model.fit(x,y)

new_size = np.array([[1600]])
predict_price= model.predict(new_size)

print(predict_price)


  
```
